```python
# Install Bokeh (only needed in Colab or Jupyter)
# !pip install bokeh  # Uncomment if running in Google Colab

import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource

# Load the dataset (ensure correct file path)
#file_path = "/mnt/data/ecommerce_product_dataset.csv"
#The original file path was likely incorrect. Update it with the correct path to the dataset.
#If the dataset is in the same directory as the notebook, you can simply use the file name:
file_path = "ecommerce_product_dataset.csv"
data = pd.read_csv(file_path)

# Compute the discounted price
data["Discounted_Price"] = data["Price"] - (data["Price"] * data["Discount"])

# Create a ColumnDataSource for Bokeh
source = ColumnDataSource(data={"Price": data["Price"], "Discounted_Price": data["Discounted_Price"]})

# Enable inline Bokeh output (for Jupyter/Colab)
output_notebook()

# Create a scatter plot
p = figure(title="Discounted Price vs. Original Price",
           x_axis_label="Original Price", y_axis_label="Discounted Price",
           width=800, height=400)

p.circle(x="Price", y="Discounted_Price", source=source, size=10, alpha=0.5, color="blue")

# Show the plot
show(p)
```
