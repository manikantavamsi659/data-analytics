```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import pandas as pd
import plotly.graph_objects as go

# Load the CSV file
# Changed the path to '/content/world_population.csv' based on the previous code
df = pd.read_csv("/content/world_population.csv")

# Normalize column names
df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]

app = dash.Dash(__name__)

app.layout = html.Div([
    dcc.Input(id='country-input', type='text', placeholder='Enter country'),
    dcc.Graph(id='population-graph'),
    html.Div(id='error-message', style={'color': 'red'})
])

@app.callback(
    [Output('population-graph', 'figure'), Output('error-message', 'children')],
    [Input('country-input', 'value')]
)
def update_graph(country):
    if not country:
        return go.Figure(), "Please enter a country."

    country = country.lower()
    country_data = df[df['country/territory'].str.lower() == country]

    if country_data.empty:
        return go.Figure(), f"No data found for {country.title()}."

    # Extract population data over the years
    years = ['2022_population', '2020_population', '2015_population', '2010_population', '2000_population',
             '1990_population', '1980_population', '1970_population']
    population_values = [country_data[year].values[0] for year in years]

    figure = go.Figure(data=[
        go.Bar(x=years, y=population_values, name='Population', marker={'color': 'blue'})
    ])

    figure.update_layout(title=f"Population Data for {country.title()}", xaxis_title="Year", yaxis_title="Population")

    return figure, ""

if __name__ == '__main__':
    app.run_server(debug=True)
```
