TASK 03 - PLOTTING THE DISTRIBUTION OF VARIOUS DISCRETE FEATURES ON SEASON/HOLIDAY/WEEKEND/WEATHERCODE


```python
# prompt:  PLOTTING THE DISTRIBUTION OF VARIOUS DISCRETE FEATURES ON SEASON/HOLIDAY/WEEKEND/WEATHERCODE

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'season', 'holiday', 'weekend', and 'weathercode' are columns in your DataFrame

def plot_discrete_feature_distributions(data, features):
    # Get the actual column names from the DataFrame
    valid_features = [col for col in features if col in data.columns]

    # Print a warning if any features were not found
    invalid_features = [col for col in features if col not in data.columns]
    if invalid_features:
        print(f"Warning: The following features were not found in the DataFrame: {invalid_features}")

    for feature in valid_features:
        plt.figure(figsize=(10, 6))
        sns.countplot(x=feature, data=data)
        plt.title(f'Distribution of {feature}')
        plt.xlabel(feature)
        plt.ylabel('Count')
        plt.show()

# Example usage:
discrete_features = ['season', 'holiday', 'weekend', 'weathercode']  # Replace with your actual feature names
plot_discrete_feature_distributions(data, discrete_features)
```

TASK 04 - LOOKING AT THE DATA TYPES OF EACH VARIABLE, TRANSFORM TIMESTAMP IN TYPE & SET IT AS INDEX


```python
import pandas as pd

# Assuming 'data' is your DataFrame and 'timestamp' is the column name
# 1. Examine data types
print(data.dtypes)

# 2. Transform 'timestamp' to datetime
data['timestamp'] = pd.to_datetime(data['timestamp'])

# 3. Set 'timestamp' as index
data = data.set_index('timestamp')

# Now, your DataFrame 'data' has 'timestamp' as its index, and the data type of 'timestamp' is datetime.
```

TASK 06 - VISUALIZING THE CORRELATION WITH A HEATMAP



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame with the features

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Plot the heatmap
plt.figure(figsize=(12, 10))  # Adjust figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

TASK 07 - VISUALIZING THE CORRELATION OF THE TARGET VARIABLE & THE OTHER FEATURES WITH BARPLO


```python
# prompt:  VISUALIZING THE CORRELATION OF THE TARGET VARIABLE & THE OTHER FEATURES WITH BARPLO

# Assuming 'data' is your DataFrame and 'cnt' is your target variable

# Calculate correlations between 'cnt' and other numerical features
correlations = data.corr()['cnt'].drop('cnt')  # Exclude 'cnt' itself

# Create a bar plot
plt.figure(figsize=(10, 6))
correlations.plot(kind='bar')
plt.title('Correlation of Features with Target Variable (cnt)')
plt.xlabel('Features')
plt.ylabel('Correlation')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

TASK 08 - PLOTTING BIKE SHARES OVER TIME BY USING LINEPLOT


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'data' is your DataFrame with 'timestamp' as the index and 'cnt' as the bike share count column

# Plot bike shares over time
plt.figure(figsize=(15, 6))  # Adjust figure size as needed
sns.lineplot(x=data.index, y='cnt', data=data)
plt.title('Bike Shares Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Bike Share Count')
plt.grid(True)  # Add grid for better readability
plt.show()
```

**TASK 09**- PLOTTING BIKE SHARES BY MONTHS & YEAR_OF_MONTH (USE LINEPLOT, POINTPOT, BARPLOT)



```python
# prompt: PLOTTING BIKE SHARES BY MONTHS & YEAR_OF_MONTH (USE LINEPLOT, POINTPOT, BARPLOT)

# TASK 09 - PLOTTING BIKE SHARES BY MONTHS & YEAR_OF_MONTH (USE LINEPLOT, POINTPLOT, BARPLOT)

# Group data by month and year
monthly_data = data.groupby(['year', 'month'])['cnt'].mean().reset_index()

# Plotting with different plot types
plt.figure(figsize=(15, 5))

# Line plot
plt.subplot(1, 3, 1)
sns.lineplot(x='month', y='cnt', hue='year', data=monthly_data)
plt.title('Monthly Bike Shares (Line Plot)')

# Point plot
plt.subplot(1, 3, 2)
sns.pointplot(x='month', y='cnt', hue='year', data=monthly_data)
plt.title('Monthly Bike Shares (Point Plot)')


# Bar plot
plt.subplot(1, 3, 3)
sns.barplot(x='month', y='cnt', hue='year', data=monthly_data)
plt.title('Monthly Bike Shares (Bar Plot)')

plt.tight_layout()  # Adjust subplot parameters for a tight layout
plt.show()
```

**TASK 14** - VISUALIZING THE DISTRIBUTION OF BIKE SHARES BY WEEKDAY-WEEKEND WITH USING PIECHART & BARPLOT

```python
# prompt: VISUALIZING THE DISTRIBUTION OF BIKE SHARES BY WEEKDAY-WEEKEND WITH USING PIECHART & BARPLOT

# Assuming 'data' is your DataFrame and it has a 'is_weekend' column (0 or 1) and 'cnt' (bike share count).

# Calculate the total bike shares for weekdays and weekends
weekday_shares = data[data['is_weekend'] == 0]['cnt'].sum()  # Changed 'weekend' to 'is_weekend'
weekend_shares = data[data['is_weekend'] == 1]['cnt'].sum()  # Changed 'weekend' to 'is_weekend'

# Create a pie chart
labels = ['Weekday', 'Weekend']
sizes = [weekday_shares, weekend_shares]
explode = (0.1, 0)  # Explode the 'Weekday' slice for emphasis
plt.figure(figsize=(6, 6))  # Adjust figure size
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Distribution of Bike Shares by Weekday/Weekend')
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle
plt.show()


# Create a bar plot
plt.figure(figsize=(6, 4))  # Adjust figure size
plt.bar(labels, sizes)
plt.xlabel('Day Type')
plt.ylabel('Total Bike Shares')
plt.title('Distribution of Bike Shares by Weekday/Weekend')
plt.show()
```
